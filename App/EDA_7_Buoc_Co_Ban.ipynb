## B∆Ø·ªöC 1: IMPORT TH∆Ø VI·ªÜN V√Ä ƒê·ªåC D·ªÆ LI·ªÜU


```python
# Import c√°c th∆∞ vi·ªán c·∫ßn thi·∫øt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Thi·∫øt l·∫≠p hi·ªÉn th·ªã
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
pd.set_option('display.width', None)

# Thi·∫øt l·∫≠p style cho bi·ªÉu ƒë·ªì
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# ƒê·ªçc d·ªØ li·ªáu
print("="*80)
print("B∆Ø·ªöC 1: IMPORT TH∆Ø VI·ªÜN V√Ä ƒê·ªåC D·ªÆ LI·ªÜU")
print("="*80)

# ƒê·ªçc file CSV
df = pd.read_csv('result.csv', encoding='utf-8')

print(f"\n‚úÖ ƒê√£ ƒë·ªçc d·ªØ li·ªáu th√†nh c√¥ng!")
print(f"   - S·ªë d√≤ng: {len(df):,}")
print(f"   - S·ªë c·ªôt: {len(df.columns)}")
```

## B∆Ø·ªöC 3: KI·ªÇM TRA D·ªÆ LI·ªÜU THI·∫æU (MISSING VALUES)


```python
print("="*80)
print("B∆Ø·ªöC 3: KI·ªÇM TRA D·ªÆ LI·ªÜU THI·∫æU")
print("="*80)

# 3.1. T·ªïng s·ªë gi√° tr·ªã thi·∫øu theo c·ªôt
missing_count = df.isnull().sum()
missing_percent = (missing_count / len(df)) * 100

# 3.2. T·∫°o b·∫£ng b√°o c√°o
missing_df = pd.DataFrame({
    'C·ªôt': missing_count.index,
    'S·ªë l∆∞·ª£ng thi·∫øu': missing_count.values,
    'T·ª∑ l·ªá (%)': missing_percent.values.round(2)
})

# 3.3. L·ªçc ch·ªâ c√°c c·ªôt c√≥ d·ªØ li·ªáu thi·∫øu
missing_df_filtered = missing_df[missing_df['S·ªë l∆∞·ª£ng thi·∫øu'] > 0].sort_values('S·ªë l∆∞·ª£ng thi·∫øu', ascending=False)

if len(missing_df_filtered) > 0:
    print(f"\n‚ö†Ô∏è PH√ÅT HI·ªÜN D·ªÆ LI·ªÜU THI·∫æU:")
    print(missing_df_filtered.to_string(index=False))
    
    # 3.4. T·ªïng s·ªë gi√° tr·ªã thi·∫øu
    total_missing = missing_count.sum()
    print(f"\nüìä T·ªîNG K·∫æT:")
    print(f"   - T·ªïng s·ªë gi√° tr·ªã thi·∫øu: {total_missing:,}")
    print(f"   - T·ª∑ l·ªá thi·∫øu tr√™n to√†n b·ªô d·ªØ li·ªáu: {(total_missing / df.size * 100):.2f}%")
    
    # 3.5. Visualize missing values
    plt.figure(figsize=(10, 6))
    missing_df_sorted = missing_df_filtered.sort_values('T·ª∑ l·ªá (%)', ascending=True)
    plt.barh(missing_df_sorted['C·ªôt'], missing_df_sorted['T·ª∑ l·ªá (%)'])
    plt.xlabel('T·ª∑ l·ªá thi·∫øu (%)')
    plt.title('Bi·ªÉu ƒë·ªì d·ªØ li·ªáu thi·∫øu theo c·ªôt')
    plt.tight_layout()
    plt.show()
else:
    print(f"\n‚úÖ KH√îNG C√ì D·ªÆ LI·ªÜU THI·∫æU!")
```

## B∆Ø·ªöC 4: KI·ªÇM TRA D·ªÆ LI·ªÜU TR√ôNG L·∫∂P (DUPLICATES)


```python
print("="*80)
print("B∆Ø·ªöC 4: KI·ªÇM TRA D·ªÆ LI·ªÜU TR√ôNG L·∫∂P")
print("="*80)

# 4.1. Ki·ªÉm tra d√≤ng tr√πng l·∫∑p ho√†n to√†n
duplicate_rows = df.duplicated()
num_duplicates = duplicate_rows.sum()

print(f"\nüìä K·∫æT QU·∫¢ KI·ªÇM TRA:")
print(f"   - S·ªë d√≤ng tr√πng l·∫∑p ho√†n to√†n: {num_duplicates:,}")
print(f"   - T·ª∑ l·ªá tr√πng l·∫∑p: {(num_duplicates / len(df) * 100):.2f}%")

if num_duplicates > 0:
    print(f"\n‚ö†Ô∏è PH√ÅT HI·ªÜN D·ªÆ LI·ªÜU TR√ôNG L·∫∂P!")
    print(f"\nüëÄ XEM M·∫™U C√ÅC D√íNG TR√ôNG L·∫∂P (10 d√≤ng ƒë·∫ßu):")
    duplicate_samples = df[duplicate_rows].head(10)
    display(duplicate_samples)
    
    # 4.2. Ki·ªÉm tra tr√πng l·∫∑p theo c·ªôt c·ª• th·ªÉ (v√≠ d·ª•: id)
    if 'id' in df.columns:
        duplicate_ids = df['id'].duplicated()
        num_duplicate_ids = duplicate_ids.sum()
        print(f"\nüìã KI·ªÇM TRA TR√ôNG L·∫∂P THEO C·ªòT 'id':")
        print(f"   - S·ªë gi√° tr·ªã id tr√πng l·∫∑p: {num_duplicate_ids:,}")
        if num_duplicate_ids > 0:
            print(f"   - L∆∞u √Ω: C√≥ th·ªÉ m·ªôt b·ªánh nh√¢n c√≥ nhi·ªÅu l·∫ßn kh√°m")
else:
    print(f"\n‚úÖ KH√îNG C√ì D·ªÆ LI·ªÜU TR√ôNG L·∫∂P HO√ÄN TO√ÄN!")
```

## B∆Ø·ªöC 5: PH√ÇN T√çCH D·ªÆ LI·ªÜU PH√ÇN LO·∫†I (CATEGORICAL DATA)


```python
print("="*80)
print("B∆Ø·ªöC 5: PH√ÇN T√çCH D·ªÆ LI·ªÜU PH√ÇN LO·∫†I")
print("="*80)

# 5.1. X√°c ƒë·ªãnh c√°c c·ªôt ph√¢n lo·∫°i (object, category)
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

print(f"\nüìã C√ÅC C·ªòT PH√ÇN LO·∫†I: {len(categorical_cols)} c·ªôt")
for col in categorical_cols:
    print(f"   - {col}")

# 5.2. Ph√¢n t√≠ch t·ª´ng c·ªôt ph√¢n lo·∫°i
for col in categorical_cols:
    print(f"\n" + "="*60)
    print(f"üìä PH√ÇN T√çCH C·ªòT: {col}")
    print("="*60)
    
    # S·ªë gi√° tr·ªã duy nh·∫•t
    unique_count = df[col].nunique()
    print(f"   - S·ªë gi√° tr·ªã duy nh·∫•t: {unique_count}")
    
    # Top 10 gi√° tr·ªã ph·ªï bi·∫øn nh·∫•t
    print(f"\n   üîù TOP 10 GI√Å TR·ªä PH·ªî BI·∫æN NH·∫§T:")
    value_counts = df[col].value_counts().head(10)
    value_percent = (df[col].value_counts(normalize=True) * 100).head(10)
    
    result_df = pd.DataFrame({
        'Gi√° tr·ªã': value_counts.index,
        'S·ªë l∆∞·ª£ng': value_counts.values,
        'T·ª∑ l·ªá (%)': value_percent.values.round(2)
    })
    print(result_df.to_string(index=False))
    
    # Ki·ªÉm tra gi√° tr·ªã null (n·∫øu c√≥)
    null_count = df[col].isnull().sum()
    if null_count > 0:
        print(f"\n   ‚ö†Ô∏è S·ªë gi√° tr·ªã null: {null_count} ({(null_count/len(df)*100):.2f}%)")
```

```python
# 5.3. Visualize ph√¢n b·ªë c·ªßa c√°c c·ªôt ph√¢n lo·∫°i quan tr·ªçng
important_categorical = []
for col in ['gioi_tinh', 'nhom_tuoi', 'trang_thai', 'loai_kham', 'nhom_mau', 'thanh_pho']:
    if col in df.columns:
        important_categorical.append(col)

if len(important_categorical) > 0:
    n_cols = min(len(important_categorical), 4)
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    axes = axes.ravel()
    
    for idx, col in enumerate(important_categorical[:4]):
        value_counts = df[col].value_counts().head(10)
        axes[idx].barh(range(len(value_counts)), value_counts.values)
        axes[idx].set_yticks(range(len(value_counts)))
        axes[idx].set_yticklabels(value_counts.index)
        axes[idx].set_xlabel('S·ªë l∆∞·ª£ng')
        axes[idx].set_title(f'Ph√¢n b·ªë: {col}')
        axes[idx].invert_yaxis()
    
    plt.tight_layout()
    plt.show()
else:
    print("Kh√¥ng t√¨m th·∫•y c√°c c·ªôt ph√¢n lo·∫°i quan tr·ªçng ƒë·ªÉ visualize")
```

## B∆Ø·ªöC 6: PH√ÇN T√çCH D·ªÆ LI·ªÜU S·ªê (NUMERICAL DATA)


```python
print("="*80)
print("B∆Ø·ªöC 6: PH√ÇN T√çCH D·ªÆ LI·ªÜU S·ªê")
print("="*80)

# 6.1. X√°c ƒë·ªãnh c√°c c·ªôt s·ªë
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()

print(f"\nüìã C√ÅC C·ªòT S·ªê: {len(numerical_cols)} c·ªôt")
for col in numerical_cols:
    print(f"   - {col}")

if len(numerical_cols) > 0:
    # 6.2. Th·ªëng k√™ m√¥ t·∫£
    print(f"\nüìä TH·ªêNG K√ä M√î T·∫¢:")
    display(df[numerical_cols].describe())
else:
    print("\n‚ö†Ô∏è Kh√¥ng c√≥ c·ªôt d·ªØ li·ªáu s·ªë trong dataset!")
```

```python
# 6.3. Ph√¢n t√≠ch chi ti·∫øt t·ª´ng c·ªôt s·ªë v√† ph√°t hi·ªán outliers
if len(numerical_cols) > 0:
    for col in numerical_cols:
        print(f"\n" + "="*60)
        print(f"üìä PH√ÇN T√çCH C·ªòT: {col}")
        print("="*60)
        
        print(f"   - Mean (Trung b√¨nh): {df[col].mean():.2f}")
        print(f"   - Median (Trung v·ªã): {df[col].median():.2f}")
        print(f"   - Std (ƒê·ªô l·ªách chu·∫©n): {df[col].std():.2f}")
        print(f"   - Min (Gi√° tr·ªã nh·ªè nh·∫•t): {df[col].min():.2f}")
        print(f"   - Max (Gi√° tr·ªã l·ªõn nh·∫•t): {df[col].max():.2f}")
        print(f"   - Range (Kho·∫£ng): {df[col].max() - df[col].min():.2f}")
        
        # Ph√°t hi·ªán outliers b·∫±ng IQR method
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        num_outliers = len(outliers)
        
        print(f"\n   üîç PH√ÅT HI·ªÜN OUTLIERS (IQR Method):")
        print(f"      - Q1 (25%): {Q1:.2f}")
        print(f"      - Q3 (75%): {Q3:.2f}")
        print(f"      - IQR: {IQR:.2f}")
        print(f"      - Lower bound: {lower_bound:.2f}")
        print(f"      - Upper bound: {upper_bound:.2f}")
        print(f"      - S·ªë outliers: {num_outliers:,} ({(num_outliers/len(df)*100):.2f}%)")
```

```python
# 6.4. Visualize ph√¢n b·ªë d·ªØ li·ªáu s·ªë
if len(numerical_cols) > 0:
    n_cols = min(len(numerical_cols), 4)
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    axes = axes.ravel()
    
    for idx, col in enumerate(numerical_cols[:4]):
        axes[idx].hist(df[col].dropna(), bins=50, edgecolor='black', alpha=0.7)
        axes[idx].set_xlabel(col)
        axes[idx].set_ylabel('T·∫ßn su·∫•t')
        axes[idx].set_title(f'Ph√¢n b·ªë: {col}')
        axes[idx].axvline(df[col].mean(), color='r', linestyle='--', label='Mean')
        axes[idx].axvline(df[col].median(), color='g', linestyle='--', label='Median')
        axes[idx].legend()
    
    plt.tight_layout()
    plt.show()
    
    # Box plot ƒë·ªÉ ph√°t hi·ªán outliers
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    axes = axes.ravel()
    
    for idx, col in enumerate(numerical_cols[:4]):
        axes[idx].boxplot(df[col].dropna())
        axes[idx].set_ylabel(col)
        axes[idx].set_title(f'Box Plot: {col}')
    
    plt.tight_layout()
    plt.show()
```

## B∆Ø·ªöC 7: PH√ÇN T√çCH M·ªêI QUAN H·ªÜ V√Ä T·ªîNG K·∫æT


```python
print("="*80)
print("B∆Ø·ªöC 7: PH√ÇN T√çCH M·ªêI QUAN H·ªÜ V√Ä T·ªîNG K·∫æT")
print("="*80)

# 7.1. Correlation matrix (n·∫øu c√≥ d·ªØ li·ªáu s·ªë)
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
if len(numerical_cols) > 1:
    print(f"\nüìä MA TR·∫¨N T∆Ø∆†NG QUAN:")
    correlation_matrix = df[numerical_cols].corr()
    display(correlation_matrix)
    
    # Visualize correlation matrix
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
                square=True, linewidths=1, cbar_kws={"shrink": 0.8}, fmt='.2f')
    plt.title('Ma tr·∫≠n t∆∞∆°ng quan gi·ªØa c√°c bi·∫øn s·ªë')
    plt.tight_layout()
    plt.show()
elif len(numerical_cols) == 1:
    print(f"\n‚ö†Ô∏è Ch·ªâ c√≥ 1 c·ªôt s·ªë, kh√¥ng th·ªÉ t·∫°o ma tr·∫≠n t∆∞∆°ng quan")
else:
    print(f"\n‚ö†Ô∏è Kh√¥ng c√≥ c·ªôt d·ªØ li·ªáu s·ªë ƒë·ªÉ ph√¢n t√≠ch t∆∞∆°ng quan")
```

```python
# 7.2. Ph√¢n t√≠ch m·ªëi quan h·ªá gi·ªØa bi·∫øn ph√¢n lo·∫°i v√† s·ªë
# V√≠ d·ª•: Ph√¢n t√≠ch tu·ªïi theo gi·ªõi t√≠nh
if 'tuoi' in df.columns and 'gioi_tinh' in df.columns:
    print(f"\nüìä PH√ÇN T√çCH TU·ªîI THEO GI·ªöI T√çNH:")
    age_by_gender = df.groupby('gioi_tinh')['tuoi'].agg(['mean', 'median', 'std', 'count'])
    display(age_by_gender)
    
    # Visualize
    plt.figure(figsize=(10, 6))
    df.boxplot(column='tuoi', by='gioi_tinh', ax=plt.gca())
    plt.title('Ph√¢n b·ªë tu·ªïi theo gi·ªõi t√≠nh')
    plt.suptitle('')  # X√≥a title m·∫∑c ƒë·ªãnh
    plt.xlabel('Gi·ªõi t√≠nh')
    plt.ylabel('Tu·ªïi')
    plt.tight_layout()
    plt.show()

# Ph√¢n t√≠ch tu·ªïi theo nh√≥m tu·ªïi
if 'tuoi' in df.columns and 'nhom_tuoi' in df.columns:
    print(f"\nüìä PH√ÇN T√çCH TU·ªîI THEO NH√ìM TU·ªîI:")
    age_by_group = df.groupby('nhom_tuoi')['tuoi'].agg(['mean', 'median', 'std', 'count'])
    display(age_by_group)
```

```python
# 7.3. T·ªïng k·∫øt EDA
print(f"\n" + "="*80)
print("üìã T·ªîNG K·∫æT EDA")
print("="*80)

print(f"\n1Ô∏è‚É£ K√çCH TH∆Ø·ªöC D·ªÆ LI·ªÜU:")
print(f"   - S·ªë d√≤ng: {df.shape[0]:,}")
print(f"   - S·ªë c·ªôt: {df.shape[1]}")

print(f"\n2Ô∏è‚É£ D·ªÆ LI·ªÜU THI·∫æU:")
missing_total = df.isnull().sum().sum()
if missing_total > 0:
    print(f"   - T·ªïng s·ªë gi√° tr·ªã thi·∫øu: {missing_total:,}")
    print(f"   - T·ª∑ l·ªá: {(missing_total / df.size * 100):.2f}%")
    print(f"   - ‚ö†Ô∏è C·∫ßn x·ª≠ l√Ω d·ªØ li·ªáu thi·∫øu")
else:
    print(f"   - ‚úÖ Kh√¥ng c√≥ d·ªØ li·ªáu thi·∫øu")

print(f"\n3Ô∏è‚É£ D·ªÆ LI·ªÜU TR√ôNG L·∫∂P:")
duplicate_count = df.duplicated().sum()
if duplicate_count > 0:
    print(f"   - S·ªë d√≤ng tr√πng l·∫∑p: {duplicate_count:,}")
    print(f"   - T·ª∑ l·ªá: {(duplicate_count / len(df) * 100):.2f}%")
    print(f"   - ‚ö†Ô∏è C·∫ßn ki·ªÉm tra v√† x·ª≠ l√Ω d·ªØ li·ªáu tr√πng l·∫∑p")
else:
    print(f"   - ‚úÖ Kh√¥ng c√≥ d·ªØ li·ªáu tr√πng l·∫∑p ho√†n to√†n")

print(f"\n4Ô∏è‚É£ LO·∫†I D·ªÆ LI·ªÜU:")
print(f"   - C·ªôt s·ªë: {len(df.select_dtypes(include=[np.number]).columns)}")
print(f"   - C·ªôt ph√¢n lo·∫°i: {len(df.select_dtypes(include=['object', 'category']).columns)}")

print(f"\n5Ô∏è‚É£ C√ÅC PH√ÅT HI·ªÜN QUAN TR·ªåNG:")
# Th√™m c√°c ph√°t hi·ªán quan tr·ªçng
if 'tuoi' in df.columns:
    print(f"   - Tu·ªïi trung b√¨nh: {df['tuoi'].mean():.1f} tu·ªïi")
    print(f"   - Tu·ªïi nh·ªè nh·∫•t: {df['tuoi'].min()} tu·ªïi")
    print(f"   - Tu·ªïi l·ªõn nh·∫•t: {df['tuoi'].max()} tu·ªïi")

if 'gioi_tinh' in df.columns:
    gender_dist = df['gioi_tinh'].value_counts()
    print(f"   - Ph√¢n b·ªë gi·ªõi t√≠nh: {dict(gender_dist)}")

if 'nhom_tuoi' in df.columns:
    age_group_dist = df['nhom_tuoi'].value_counts()
    print(f"   - Ph√¢n b·ªë nh√≥m tu·ªïi: {dict(age_group_dist.head())}")

print(f"\n6Ô∏è‚É£ ƒê·ªÄ XU·∫§T:")
print(f"   - Ti·∫øp t·ª•c l√†m s·∫°ch d·ªØ li·ªáu n·∫øu c·∫ßn")
print(f"   - Feature engineering cho c√°c bi·∫øn ph√¢n lo·∫°i")
print(f"   - X·ª≠ l√Ω outliers n·∫øu c·∫ßn thi·∫øt")
print(f"   - Chu·∫©n b·ªã d·ªØ li·ªáu cho modeling")

print(f"\n‚úÖ HO√ÄN TH√ÄNH EDA!")
print("="*80)
```
